```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Load your dataset
data = pd.read_csv("career_data_large.csv")



# Initialize OneHotEncoder for the "CareerOutcome" column
onehot_encoder = OneHotEncoder(sparse=False)

# Encode the "CareerOutcome" column using one-hot encoding
career_encoded = onehot_encoder.fit_transform(data[['CareerOutcome']])

# Convert the one-hot encoded career outcomes into a DataFrame
career_encoded_df = pd.DataFrame(career_encoded, columns=onehot_encoder.get_feature_names(['CareerOutcome']))

# Drop the original "CareerOutcome" column since it's now one-hot encoded
data.drop(columns=['CareerOutcome'], inplace=True)

# Concatenate the one-hot encoded columns with the original data
data_encoded = pd.concat([data, career_encoded_df], axis=1)

# Display the first few rows of the encoded dataset
print(data_encoded.head())

# Optionally, you can save the encoded dataset to a new CSV file
# data_encoded.to_csv("encoded_career_data.csv", index=False)


# Initialize OneHotEncoder for the "Stream" column
onehot_encoder = OneHotEncoder()

# Encode the "Stream" column using one-hot encoding
stream_encoded = onehot_encoder.fit_transform(data[['Stream']])

# Convert the one-hot encoded stream into a DataFrame and concatenate with the original data
stream_encoded_df = pd.DataFrame(stream_encoded.toarray(), columns=onehot_encoder.get_feature_names(['Stream']))
data_encoded = pd.concat([data, stream_encoded_df,career_encoded_df], axis=1)

# Drop the original "Stream" column since it's now one-hot encoded
data_encoded.drop(columns=['Stream'], inplace=True)

# Display the first few rows of the encoded dataset
print(data_encoded.head())

# Optionally, you can save the encoded dataset to a new CSV file
# data_encoded.to_csv("encoded_career_data.csv", index=False)
```

```python
data_encoded.head()
```

```python
X = data_encoded.iloc[:, :4].values
y = data_encoded.iloc[:, 4:].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Assuming you have predictions from your linear regression model in 'y_pred' and true values in 'y_true'
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Print the regression metrics
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```
